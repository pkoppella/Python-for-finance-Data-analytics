```python
names = ['sara','nav','rich','pal']
names
```

```python
random_names = [names[random.randint(low=0,high=len(names))] for i in range(1000)]
```

```python
random_names[:10]
```

```python
births = [random.randint(low=0,high=89) for i in range(1000)]
births[:10]
```

```python
#now merge both births and birth_names
BabySet = list(zip(random_names,births))
BabySet[:10]
```

```python
df = pd.DataFrame(data = BabySet , columns= ['names','births'])
df[:10]
```

```python
df.to_csv('/Users/rgottumukkal/jobs/births1880.csv')
print(df.describe())
```
